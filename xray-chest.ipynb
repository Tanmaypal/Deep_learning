```python
import keras
from keras.models import Model
from keras import applications
from keras.layers import GlobalAveragePooling2D,Dropout,Dense
!pip install livelossplot
from livelossplot import PlotLossesKeras
```

```python
from keras.preprocessing.image import ImageDataGenerator
import numpy as np
import matplotlib.pyplot as plt
train_path = '../input/chest-xray-pneumonia/chest_xray/train'
#test_path = '/content/cell_images'
batch_size = 128
image_size = 224
num_class = 2
EPOCHS=15
```

```python
train_datagen = ImageDataGenerator(
                                   shear_range=0.2,
                                   zoom_range=0.2,
                                   rescale=1.0/255,
                                   horizontal_flip=True)
```

```python
train_generator = train_datagen.flow_from_directory(
                        directory=train_path,
                        target_size=(image_size,image_size),
                        batch_size=batch_size,
                        subset='training',
                        class_mode='categorical',
                        shuffle=True)
```

```python
validation_generator = train_datagen.flow_from_directory(
                        directory='../input/chest-xray-pneumonia/chest_xray/val/',
                        target_size=(image_size,image_size),
                        batch_size=batch_size,
                        class_mode='categorical',
                        shuffle=True)
```

```python
test_data_generator = ImageDataGenerator(
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    vertical_flip=True)
```

```python
#test_data_dir='/kaggle/TEST/'
test_generator = test_data_generator.flow_from_directory(
    directory='/kaggle/TEST/',
    target_size=(224,224),
    batch_size=1,
    class_mode="binary", 
    shuffle=False)
```

```python
x_batch, y_batch = train_generator.next()
fig=plt.figure()
columns = 4
rows = 4
for i in range(1, columns*rows):
    num = np.random.randint(batch_size)
    image = x_batch[num].astype(np.int)
    fig.add_subplot(rows, columns, i)
    plt.imshow(image)
plt.show()
```

```python
base_model = applications.resnet50.ResNet50(weights='imagenet', include_top=False, input_shape= (224,224,3))
```

```python
x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dropout(0.7)(x)
predictions = Dense(2, activation= 'softmax')(x)
model = Model(inputs = base_model.input, outputs = predictions)###Thsi is used to fit.
```

```python
from keras.optimizers import SGD, Adam
adam = Adam(lr=0.0001)
model.compile(optimizer= adam, loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
print(model.summary())
```

```python
model.fit_generator(
    train_generator,
    steps_per_epoch=len(train_generator.filenames) // batch_size, ##Number of Batches
    epochs=EPOCHS,
    validation_data=validation_generator,
    validation_steps=len(validation_generator.filenames) //batch_size,
    callbacks=[PlotLossesKeras()], 
    verbose=1)
```
